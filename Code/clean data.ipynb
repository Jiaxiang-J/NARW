```python
from pykrige.ok import OrdinaryKriging
def kriging(lon_,lat_,var,lon,lat):
    df = OrdinaryKriging(x=lon_,y=lat_,z=var,variogram_model='spherical',verbose=False,variogram_parameters={'sill': 1.0, 'range': 0.5, 'nugget': 0.1})
    a,b = df.execute('points',lon,lat)
    return a   
```

```python
glider_dataframes = {
    '20_07': glider_20_07,
    '20_10': glider_20_10,
    '20_11': glider_20_11,
    '21_02': glider_21_02,
    '21_11': glider_21_11,
    '22_01': glider_22_01,
    '22_02': glider_22_02,
    '22_03': glider_22_03,
    '22_05': glider_22_05,
    '23_05': glider_23_05,
    '23_06': glider_23_06,
    '23_08': glider_23_08
}

# column needs to interpolate
columns_to_clean = ['chl_oc3', 'sst','M_WK','M_WK_G']
lon_total = {}
lat_total = {}

cleaned_dataframes = {}
# save after drop NA 
lon_dataframes = {}
lat_dataframes = {}
val_dataframes = {}

for date, df in glider_dataframes.items():
    # iterate columns need to clean
    for column in columns_to_clean:
        # key name
        key_name = f'{column}_{date}'
        lon_total[key_name] = df['lon']
        lat_total[key_name] = df['lat']
        
        cleaned_df = df.dropna(subset=[column])
        cleaned_dataframes[key_name] = cleaned_df
        lat_dataframes[key_name] = cleaned_df['lat']
        lon_dataframes[key_name] = cleaned_df['lon']
        val_dataframes[key_name] = cleaned_df[column]        
```

```python
## interpolate
columns_to_interpolate = ['chl_oc3','sst']
kriging_results = {}
for date,df in glider_dataframes.items():
    
    for column in columns_to_interpolate:
        key_name = f'{column}_{date}'
        kriging_results[key_name] = kriging(lon_dataframes[key_name],lat_dataframes[key_name],
                                            val_dataframes[key_name],lon_total[key_name],lat_total[key_name])
```
